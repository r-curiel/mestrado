```python
import pandas as pd
import os
```

```python
root = os.getcwd()
```

```python
df = pd.read_csv(root + '/base_dados/distancias_matriz_long.csv')
```

```python
df.DISTANCIA.fillna(0, inplace=True)
```

```python
df
```

```python
centros = pd.read_csv(root + '/base_dados/centros.csv', sep=';')
centros_list = centros.cen_ibge
centros_list
```

```python
# filter df for, if centros_list in DE or PARA, then DISTACIA <= than 200, else, the DISTANCIA <= 100
analise_centros =  df[(df['DE'].isin(centros_list)) | (df['PARA'].isin(centros_list))]
analise_centros = analise_centros[analise_centros.DISTANCIA <= 200]
# > 0
analise_centros = analise_centros[analise_centros.DISTANCIA > 0]

analise_centros
```

```python
# remove rows from df where its index existis in analise_centros <= 100 and > 0
analise_cidades = df[~df.index.isin(analise_centros.index)]
analise_cidades = analise_cidades[analise_cidades.DISTANCIA <= 100]
# > 0
analise_cidades = analise_cidades[analise_cidades.DISTANCIA > 0]
analise_cidades
```

```python
# concat both dataframes
analise = pd.concat([analise_cidades, analise_centros])
```

```python
analise.to_csv(root + '/base_dados/distancias_matriz_long_filtrado.csv', index=False, sep=';', decimal=',')
```
